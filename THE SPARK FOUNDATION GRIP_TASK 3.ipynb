# 4. Exploratory Data Analysis and Visulization

## 4.1 Total Sales and Profits of the company

```python
Sales_and_Profits = Super_Store_Data.groupby("Segment").sum().iloc[:,[1,-1]].sum()
round(Sales_and_Profits,2)
```

## 4.2 Sales and Profits of top 10 states

```python
Top_10_Sales = Super_Store_Data.groupby("State").Sales.sum().nlargest(n =10)
Top_10_Profits = Super_Store_Data.groupby("State").Profit.sum().nlargest(n =10)
```

```python
Top_10_Sales.index
```

```python
Top_10_Profits.index
```

```python
plt.style.use('seaborn-dark-palette')
Top_10_Sales.plot(kind ='bar', figsize =(14,8), fontsize =14)
plt.xlabel("States", fontsize =13)
plt.ylabel("Total Sales",fontsize =13)
plt.title("Top 10 States by Sales",fontsize =16)
plt.show()
```

```python
plt.style.use('seaborn-dark-palette')
Top_10_Profits.plot(kind ='bar', figsize =(14,8), fontsize =14)
plt.xlabel("States", fontsize =13)
plt.ylabel("Total Profits",fontsize =13)
plt.title("Top 10 States by Profits",fontsize =16)
plt.show()
```

## California and New York are far ahead of their competitor states.

## 4.3 Interdependency of Sales, Profit and Discount

```python
plt.style.use('seaborn-dark-palette')
Super_Store_Data.plot(kind = "scatter", figsize = (15,8), x = "Sales", y ="Profit" ,c ="Discount", s =20, marker ="x",colormap ="viridis")
plt.ylabel("Total Profits",fontsize =13)
plt.title("Interdependence of Sales, Profit, and Discount",fontsize =16)
plt.show()
```

## Relation Analysis

```python
financial=Super_Store_Data.loc[:,['Sales','Quantity','Discount','Profit']]
sns.pairplot(financial)
```


As we can see above that there is having some relation between sales and profit and also there is having some relation between Discount and Profit. Now To see what exact relation between those entities are we have to plot the heat_map. so we get more clearity

```python
correlation=financial.corr()
sns.heatmap(correlation,xticklabels=correlation.columns,yticklabels=correlation.columns,annot=True,cmap="cubehelix")
```

Now from above map we can investigate that:
1)sales and profits are positively correlated 2)Discount and Profits are Negatively correlated

Hence we consider these cases and proceed further for other investigation.

## Case-1 When Discount is equal to 0

```python
data=Super_Store_Data[Super_Store_Data['Discount']==0]
sns.relplot(x='Sales',y='Profit',data=data)
```

Hence we can say that there is having positive relation between Profit and sales.when Discount is 0 Now we plot heat_map to get correlaton.

```python
correlation=data.corr()
sns.heatmap(correlation,xticklabels=correlation.columns,yticklabels=correlation.columns,annot=True,cmap="cubehelix")
```

Hence,As we can observe that There is strong correlation between sales and Profit i.e 0.92

```python
#Rel_plot with respect category
sns.relplot(x='Sales',y='Profit',hue='Category',data=data) 
```

```python
# Regression Plot 
plt.figure(figsize=(12,6))
sns.regplot(data['Sales'],data['Profit'],color="r")
```

It is showing that there is positive trend between Profit and sales.

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='Category',y='Profit',data=data,palette="hls")
```

## Case-2 When Discount is Not equal to 0

```python
data1=Super_Store_Data[Super_Store_Data['Discount']!=0]
sns.relplot(x='Sales',y='Profit',hue='Discount',data=data1)
```

As we can see in above graph that as percentages of Discount increses the sales is also goes increses but profit goes decreases.

Now we have to check how it can be effects on different sectors of businesses.

```python
plt.figure(figsize=(12,6))
sns.relplot(x='Sales',y='Profit',hue='Category',data=data1)
```

```python
#realtion analysis
plt.figure(figsize=(12,6))
correlation=data1.corr()
sns.heatmap(correlation,xticklabels=correlation.columns,yticklabels=correlation.columns,annot=True,cmap="cubehelix")
```

```python
plt.figure(figsize=(12,8))
pivot=pd.pivot_table(data1,index='Sub-Category',values='Profit')
pivot.plot(kind='bar')
```

Here we can see that copiers had highest Profit and Machines had highest loss

```python
pivot=pd.pivot_table(data1,index='Sub-Category',values='Sales')
pivot.plot(kind='bar')
```

Here we can see that copiers had highest sale and Machines had second highest sales.

As we can observe from above two graph that 'Machines' had second highest sale but due to large discount it is in loss and in second graph we see that sales in 'Fasteners','labels'and 'Art' category are so weak.so we have to concentrate on these sub-category businesses

# Discounts are triggering losses

## 4.4 Distrubution of Profits across diffrent regions

```python
plt.figure(figsize = (12,6))
sns.set(font_scale=1, palette= "viridis")
sns.barplot(data = Super_Store_Data , x = "Region",y = "Profit" ,hue = "Segment")
plt.show()
```

## Overall each Segment is profitable.

## 4.5 Profit distribution by Region

```python
plt.figure(figsize = (12,6))
sns.set(font_scale=1, palette= "viridis")
sns.barplot(data = Super_Store_Data , x = "Region",y = "Profit" ,hue = "Category")
plt.show()
```

## As we can see that only 'furniture' sector in the central region is making loss.

## 4.6 Investigating losses in Furniture category in the Central region

```python
# Grouping Data by Region and only slicing Data for Central Region from whole Data Set
gb_Central = list(Super_Store_Data.groupby("Region"))[0][1]
```

```python
# Investing Further in cenral Region 
plt.figure(figsize = (12,6))
sns.set(font_scale=1.5, palette= "viridis")
sns.barplot(data = gb_Central, x = "Category",y = "Profit" ,hue = "Ship Mode")
plt.title("Investigation of central region: Profit making(by Ship Mode)")
plt.show()
```

## Losses are inccured in Furniture Cateory irrespective to ship mode in Central Region

```python
# Slicing Furniture Data from whole data set
gb_Category_Furniture =list(list(Super_Store_Data.groupby("Region"))[0][1].groupby("Category"))[0][1]
```

```python
# Correlation matrix Heat Map to identify key factors influening profits
plt.figure(figsize = (12,8))
sns.set(font_scale=1.4)
sns.heatmap(gb_Category_Furniture.corr() , annot = True, cmap ="rocket")
plt.show()
```

## There is unusually high positive correlation between Postal Code and Discount

## Also, Their is negative correlation between Discount and Sales eventhough dicounts are entered as positive values... i.e. they are not helping in improving sales of "Furniture" category of the company

## 4.7 Investigating individual performance by states in the central region

```python
plt.figure(figsize = (12,8))
sns.set(font_scale=1, palette= "viridis")
sns.barplot(data = gb_Category_Furniture , x = "State",y = "Profit" ,hue = "Sub-Category")
plt.title("Investigation of Central Region Furniture Category: Profit Analysis(by Sub Category)", fontsize = 20)
plt.show()
```

```python
plt.figure(figsize = (12,8))
sns.set(font_scale=1, palette= "viridis")
sns.barplot(data = gb_Category_Furniture , x = "State",y = "Profit" ,hue = "Segment")
plt.title("Investigation of Central Region Furniture Category: Profit Analysis(by Segment)", fontsize = 20)
plt.show()
```

## Texas and Illiois are only two states contributing to all the losses in Furniture category in the Central Region

## Losses in Tables Sub Category is significanlty high.

## 4.7 So, what is it they are doing diffrently?

```python
plt.figure(figsize = (12,8))
sns.set(font_scale=1, palette= "viridis")
sns.barplot(data = gb_Category_Furniture , x = "State",y = "Discount" ,hue = "Sub-Category")
plt.title("Discounts provided by each state", fontsize = 20)
plt.show()
```

## Texas and Illinois are only states providing discounts in the whole central region this justifies high positive correlation between postal codes and discounts.

## Also, these discounts are very high!

## 1. 60% on Furnishings
## 2. 30% on Bookcases and Chairs
## 3. 50% disount on Tables in Illinois and 30% in Texas

# Are these discounts driving sales up?

```python
plt.figure(figsize = (12,8))
sns.set(font_scale=1.5)
sns.lmplot(data = gb_Category_Furniture , x = "Discount", y ="Sales", aspect = 1, height = 8, col ="Sub-Category", col_wrap= 2)
plt.show()
```

```python
# Jiont plot for studying overall relationship between Sales and Discounts
sns.set(font_scale=1.5)
sns.jointplot(data = gb_Category_Furniture , x = "Discount", y ="Sales", height = 8, kind = "reg")
plt.show()
```

## Actually Sales tend to do down when discounts go up in 3 out of 4 "Subcategories" and also in overall Furniture Category Sales in the Central Region.
